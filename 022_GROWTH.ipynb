# Growth 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exponential growth
nsteps = 100             #Number of steps in our lil simulation
x = np.zeros(nsteps)     #Number of anumals at every step
x[0] = 0.1               #The original number of animals
r = 1.1                  #Rate of the popuation growth

for i in range(1,nsteps):
    x[i] = x[i-1]*r  
    plt.plot(x);
```

```python
# Logistical growth

nsteps = 100             #Number of steps in our lil simulation
x = np.zeros(nsteps)     #Number of anumals at every step
x[0] = 0.6               #The original number of animals
r = 3                  #Rate of the popuation growth

for i in range(1,nsteps):
    x[i] = r*x[i-1]*(1-x[i-1]) 
plt.plot(x);
plt.ylim((0,1)) #Force your axis to be between 0 and 1
```
