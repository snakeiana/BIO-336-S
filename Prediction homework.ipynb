## Prediction horizon
* Using the calculation to predict grass growth in relation to time.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100              # Number of steps 
ntraces = 10
x = np.zeros(shape=(nsteps,ntraces))    # 
wiggle = 0.001                          # The standard deviation of the original noise in x0

r = np.arange(start = 0, stop = 4, step = 0.2) # I wanted to make a range for but that'a not the solution.
for itrace in range(ntraces):
    x[0,itrace] = 0.6+np.random.randn(1)*wiggle
    for istep in range(1,nsteps):
        x[istep,itrace] = r*x[istep-1,itrace]*(1-x[istep-1,itrace])

fig = plt.figure(figsize=(10,5))
plt.plot(np.std(x,axis=1))
plt.xlabel('Time');
plt.ylabel('Grass height in cm');
```

```python
nsteps = 100
ntrials = 1000
x = np.zeros(shape=(nsteps,2))

diff = np.zeros(shape=(nsteps,ntrials))
wiggle = 0.001
#r = 3.9

figb = plt.figure(figsize=(10,5))

starting_points = [0.000,0.001]

for r in np.arange(start = 3.6, stop = 4, step=0.1):
    for itrials in range(ntrials):
        starting_points= np.random.uniform(low=0,high=1-wiggle)+np.array([0,wiggle])
        for ipair in range(2):
            x[0,ipair]=starting_points[ipair]
            for istep in range(1,nsteps):
                x[istep,ipair]=r*x[istep-1,ipair]*(1-x[istep-1,ipair])
        diff[:,itrials]=np.abs(x[:,1]-x[:,0])
    plt.plot(np.mean(diff,axis=1),label=np.around(r,decimals=2))
    plt.legend(loc='lower right')
plt.xlabel('step')
plt.ylabel('inability to predict')
```
