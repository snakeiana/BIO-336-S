# Simple numerical integration for motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Attempt number 1 throwing stone upwards

a = -1  #Acceleration; negative because it tires to reverse the flight of the stone

v = 5     #We throw the stone upwards; positive
x = 0     #We throw it from the ground so x is 0
dt = 0.01 #Our step, numerical integration step, is supposed to be small
nsteps = 1000
history = np.zeros(nsteps)
for i in range(nsteps):
    v = v + a*dt
    x = max(x + v*dt , 0)
    history[i] = x 
plt.plot(history);
    
```

```python
#Attempt number 2 throwing it at an angle

a = -1  #Acceleration; negative because it tires to reverse the flight of the stone
vy = 5
vx = 1
x = 0 
y = 0     
dt = 0.01 
nsteps = 1000
history = np.zeros(nsteps)
for i in range(nsteps):
    vy = vy + a*dt
    x = x + vx*dt
    y = max(y + vy*dt , 0)
    plt.plot(x,y,'k.'); 
```

```python
 #Attempt #2, throwing it at an angle

a = -1        # Acceleratiion; negative coz it tries to reverse the flight of the stone
vy = 5        # Vertical component of the speed, and it's upwards => positive
vx = 1        # Horizontal component of the speed
x = 0         # We throw it from our house, and are trying to see how far it goes
y = 0         # We throw it from the surface, so y=0
f = 0.02
dt = 0.05     # Our step. Numerical integration step. Is supposed to be small.
nsteps = round(10/dt)

for i in range(nsteps):
    vy = vy + a*dt - (vy*f)        # Update vertical speed
    vx = vx - vx*f
    
    x = x + vx*dt           # Update x position
    y = max(y + vy*dt , 0)  # Update y position
    
    plt.plot(x,y,'k.');      # Just plot a new point, instead of remembering x and y, coz we're lazy right now
```
