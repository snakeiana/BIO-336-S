```python
nsteps = 100
x = 0.67
r = 3
i = 0
fig = plt.figure(figsize=(9,9))
for r in np.arange(2.8,4,0.001):
    i = i + 1
    for b in range(nsteps + i%50):
        x = x*r*(1-x)
    plt.plot(r,x,'k.',ms=4)
```

# Sensitivity plot
Role-Playing: Let's predict someting with this equation.
How far in the future can we predict? 
*If the question is about time, the answer may be a plot with time as the x axis.
*If we have a plot with time as an axis and the question is about prediction quality then the y axis should come kind of prediction quality.

## How to measure prediction quality
*Use pairs of trajectories, calculate mean (abs(difference)) for each point in time

*Use a family of trajectories ; noisify (wiggle) 

```python
nsteps = 100
ntraces = 10
x = np.zeros(shape=(nsteps,ntraces))
r = 3.9
x[0]=0.66
noise = 0.0001

for itrace in range(ntraces):
    x[0,itrace]=0.6+np.random.randn(1)*noise
    for istep in range(1,nsteps):
        x[istep,itrace]=r*x[istep-1,itrace]*(1-x[istep-1,itrace])

fig = plt.figure(figsize=(10,5))
plt.plot(x)
plt.ylim((0,1))
plt.xlabel('Time')
plt.ylabel('All trajectories')

fig = plt.figure(figsize=(10,5))
plt.plot(np.std(x,axis=1))
```
